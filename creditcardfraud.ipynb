```python
X = data.iloc[:, data.columns.values !='Class']
```

```python
y = data.iloc[:, data.columns.values =='Class']
```

```python
cols = X.columns.values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier()
model_RF = RF.fit(x_train, y_train)
```

```python
y_pred = model_RF.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
n_estimators=[5,10,15]
criterion=['gini', 'entropy']
max_depth=[5,10,15]
min_samples_split=[2,4,6]
min_samples_leaf=[1,2,3]
max_features=['auto', 'sqrt', 'log2', None]
max_leaf_nodes=[3,6,9]
bootstrap=[True]
```

```python
grid_param = {'n_estimators':n_estimators, 'criterion':criterion, 'max_depth':max_depth,'min_samples_split':min_samples_split,'min_samples_leaf':min_samples_leaf,'max_features':max_features,'max_leaf_nodes':max_leaf_nodes,'bootstrap':bootstrap}
```

```python
GS = GridSearchCV(estimator=RF, param_grid = grid_param, scoring ='accuracy', n_jobs = -1, cv =10)
GS.fit(x_train, y_train)
```

```python
GS.best_estimator_
```

```python
best_model = GS.best_estimator_
best_model.fit(x_train,y_train)
y_pred = best_model.predict(x_test)
print(classification_report(y_pred, y_test))
```
