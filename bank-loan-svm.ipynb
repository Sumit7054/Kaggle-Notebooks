```python
del data1['Fiscal Year End Date'], data1['Loans'], data1['Loan Fund Sources'], data1['Recipient State'], data1['Recipient Postal Code'], data1['Date Loan Awarded']
```

```python
data1['Jobs Planned'] = data1.groupby(['Recipient Name'], sort=False)['Jobs Planned'].apply(lambda x: x.fillna(x.mean()))
```

```python
data1['Jobs Created'] = data1.groupby(['Recipient Name'], sort=False)['Jobs Created'].apply(lambda x: x.fillna(x.mean()))
```

```python
data1.isnull().sum()
```

```python
data1['Jobs Planned'].ffill(inplace=True)
```

```python
data1['Jobs Created'].ffill(inplace=True)
```

```python
data1.isnull().sum()
```

```python
data1.dropna(inplace=True)
```

```python
data1.isnull().sum()
```

```python
data1.shape
```

```python
data1.info()
```

```python
def frequency_encode(data, col):
    enc_nom_1 = (data.groupby(col).size()) / len(data)
    freqen = "Freq_{}".format(col)
    data[freqen] = data[col].apply(lambda x : enc_nom_1[x])
    del data[col]
    return data
```

```python
data1.columns
```

```python
cols = ['Authority Name', 'Recipient Name', 'Recipient City','Loan Purpose']
```

```python
data1['Loan Terms Completed'] = data1['Loan Terms Completed'].apply(lambda x: 0 if x == "No" else 1)
data1['New Jobs'] = data1['New Jobs'].apply(lambda x: 0 if x == "No" else 1)
```

```python
for col in cols:
    data1 = frequency_encode(data1, col)
```

```python
data1.info()
```

```python
data1['New Jobs'].unique()
```

```python
y = data1['Loan Terms Completed']
```

```python
del data1['Loan Terms Completed']
```

```python
x = data1
```

```python
x.info()
```

```python
from sklearn.preprocessing import StandardScaler
x = StandardScaler().fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='rbf')
svc.fit(x_train, y_train)
```

```python
y_pred = svc.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
